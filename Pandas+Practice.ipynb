Compute the average and max price charged.  Store the results in variables called avg_price and max_price

```python
#Write your code here

avg_price = df_data.price.sum()/len(df_data)
print(f'The average price: {avg_price}')
max_price = df_data.price.max()
print(f'The max price: {max_price}')
```

Compute the average price and average number of items purchased with only a single call to the mean() method. Store these averages in avg_price and avg_num_items.

```python
#Write you code here
avg_price = df_data.price.mean()
print(f'The Average price: {avg_price}')
avg_num_items = df_data['num_item'].mean()
print(f'The Average number of items: {avg_num_items}')
```

Create a column called "total", which is the product of the price and num_item columns.

```python
#Write your code here

df_data['Total'] = df_data['price'] * df_data['num_item']
df_data.head(20)
```

What is the avgerage price of the three most expensive purchases? 

```python
#Write your code here

top_3_purchases_index = df_data['Total'].sort_values(ascending=False).index[:3]

average_top_3_purchases = df_data['Total'].loc[top_3_purchases_index].mean() # If you want the average of the total purchase
print(f'The average value of the top three purchases: {average_top_3_purchases}')


average_top_3_price = df_data['price'].loc[top_3_purchases_index].mean() # If you want the average price of the products that are placed in the top three highest cost purchases
print(f'The average price of the top three purchased products: {average_top_3_price}')
```

How many different products are there?

```python
#Write your code here

num_of_products = len(df_data.product_id.unique())
print(f'There are {num_of_products} products.')
```

Find the difference between the mean and median of the column total.  Use the describe() method and then index the dataframe it returns.

```python
#Write your code here
total_mean = df_data.Total.describe()['mean']
print(f'mean: {total_mean}')

total_median = df_data.Total.describe()['50%']
print(f'median: {total_median}')

diff = total_mean - total_median
print(f'The Difference between mean and median: {diff}')
```

Use the rename method to change the column "Name" to "Passenger_Name" and the column "Ticket" to "Ticket_Num".

Change the index of the df_titanic to the column PasengerID.

```python
#Write your code here
df_titanic.reset_index(inplace=True)
df_titanic.drop(columns='index',inplace=True)
df_titanic.head(20)
```

How many missing entries are there in the Age column? Use the isnull() method and recall than Booleans are exactly like 1s (True) and 0s (False).

```python
#Write you code here
missing_ages = df_titanic.Age.isna().sum()
print(f'there are {missing_ages} missing ages.')

missing_ages_series = df_titanic.Age.isnull()
```

Compute the avg age of passengers ignoring the missing data.

```python
#Write your code here

avg_pass_age = df_titanic.Age.mean(skipna=True)
print(f'The average passenger age: {avg_pass_age}')
```

Using the fillna() method replace the missing values in the Age column with the mean.

```python
#Write your code here

df_titanic.Age.fillna(df_titanic.Age.mean(skipna=True),inplace=True)
df_titanic.Age.isna().sum()
```

What is the average age of the 5 oldest passengers? The reset_index method will be helpful here.

```python
#Write your code here
df_titanic
```

```python
age_df = df_titanic.reset_index(drop=True)
age_df.set_index('Age',inplace=True,drop=False)
age_df.sort_index(inplace=True,ascending=False)
```

```python
age_df.index.name = None
top_5_oldest = age_df.Age.iloc[:5]
top_5_oldest_mean = top_5_oldest.mean()
print(f'The average age of the top 5 passengers: {top_5_oldest_mean}')
```
